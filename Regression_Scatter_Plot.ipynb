```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
df_sales = pd.read_csv('scatter_plot_ii.csv')
```

```
df_sales
```

```
#plt.figure(figsize = (12,8))
#plt.scatter(df_real_estate['size'],
#            df_real_estate['price-k'])
#plt.show()
```

```
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_sales)
plt.show()
```

```
#
# Increase Plot Size
#
plt.figure(figsize = (12,8))
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_sales)
plt.show()
```

```
#
# Increase Plot Size using Seaborn
#
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_sales)
plt.show()
```

```
#
# Changing colors 
#  Black points
#  Red trend line
#
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_sales,
            scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'})
plt.show()
```

```
#
# Add title and labels
#
sns.set(rc = {'figure.figsize': (10,8)})
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_sales,
            scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'})
plt.xlabel('AD Expenditure in $1,000s')
plt.ylabel('Sales on Units in 1,000s')
plt.title('Effect of AD Expenditure on Sales', fontsize = 14, weight = 'bold')
plt.show()
```

```
#
# Use lmplot (google the difference between regplot and lmplot)
#
sns.lmplot(x = 'Budget',
           y = 'Sales',
           data = df_sales,
           scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'},
           height = 10)  # Height sets both Height and Width attributes
plt.xlabel('AD Expenditure in $1,000s')
plt.ylabel('Sales on Units in 1,000s')
plt.title('Effect of AD Expenditure on Sales', fontsize = 14, weight = 'bold')
plt.show()
```
