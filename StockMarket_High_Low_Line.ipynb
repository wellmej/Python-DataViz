```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
df_stock_market = pd.read_csv('HistoricalQuotes2.csv')
```

```
df_stock_market
```

```
#
# Convert date values to date-time format (yyyy-mm-dd)
#
df_stock_market["new_date"] = pd.to_datetime(df_stock_market["Date"])
```

```
df_stock_market["new_date"]
```

```
df_stock_market[" High"]
```

```
labels = ['High', 'Low']
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market["new_date"], df_stock_market[' High'])
plt.plot(df_stock_market["new_date"], df_stock_market[' Low'])
plt.title('Historical Stock Market', fontsize = 14, fontweight = 'bold')
plt.ylabel("Market Value")
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```
df_stock_market_subset = df_stock_market[(df_stock_market.new_date >= '2020-07-01') & (df_stock_market.new_date <= '2020-08-30') ]
```

```
df_stock_market_subset
```

```
labels = ['High', 'Low']

plt.figure(figsize = (20, 16))
plt.plot(df_stock_market_subset["new_date"], df_stock_market_subset[' High'], label = 'High', color = 'midnightblue')
plt.plot(df_stock_market_subset["new_date"], df_stock_market_subset[' Low'], label = 'Low', color='crimson')

plt.title('Historical Stock Market', fontsize = 14, fontweight = 'bold')
plt.ylabel("Market Value")
plt.xlabel('Date')
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```
# line 1 points
x1 = [10,25,30]
y1 = [20,40,10]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")
# line 2 points
x2 = [10,20,35]
y2 = [40,10,30]
# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Two or more lines on same plot with suitable legends ')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
